
# CAGR del precio de Bitcoin (2014 - 2021)

## Datos utilizados

- Fecha inicial: 1 de octubre de 2014  
- Precio inicial: **386.94 USD**  

- Fecha final: 1 de diciembre de 2021  
- Precio final: **57,005.43 USD**  

- Periodo analizado: **7.17 años**  

---

## Fórmula del CAGR

\[
CAGR = \left(\frac{Valor\ Final}{Valor\ Inicial}\right)^{\frac{1}{n}} - 1
\]

Donde:  
- \(n\) = número de años

Aplicando los valores:  

\[
CAGR = \left(\frac{57005.43}{386.94}\right)^{\frac{1}{7.17}} - 1
\]

Resultado:  

\[
CAGR \approx 1.00701 \Rightarrow 100.70\% \ anual
\]

---

## Interpretación

El CAGR de Bitcoin entre octubre 2014 y diciembre 2021 fue aproximadamente:  

# \(\boxed{100.70\%\ anual}\)

Esto significa que, en promedio, Bitcoin duplicó su valor cada año durante este período.  

Una rentabilidad espectacular y muy poco común en mercados tradicionales.
